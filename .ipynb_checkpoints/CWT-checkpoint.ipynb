```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from scipy import signal
```

```python
sig = np.random.rand(20) - 0.5
wavelet = signal.ricker
widths = np.arange(1, 11)
cwtmatr = signal.cwt(sig, wavelet, widths)
```

```python
plt.plot(cwtmatr)
plt.show()
```

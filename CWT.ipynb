```python
def tf(signal, fs, w=256, ylim=None):
    '''Plot a 1D signal and its spectrogram'''
    dt = 1./fs
    n = signal.size
    t = np.arange(0.0, n*dt, dt)
    noverlap = w - 1

    plt.figure(figsize=(12,8))
    plt.subplot(211)
    plt.plot(t, signal)
    ax1 = plt.subplot(212)
    Pxx, freqs, bins, im = plt.specgram(signal, NFFT=w, Fs=fs, noverlap=noverlap, cmap='cubehelix_r')
    if ylim:
        ax1.set_ylim((0,ylim))
        
    plt.show()
```

- DOG4       : 4th Derivative Of Gaussian

- HaarW      : Normalised Haar

Let's wrap everything:

```python
def cwt(data):
    
    w = pycwt.Morlet(data, largestscale=4, notes=16, order=2, scaling="log")
    
    scales = w.getscales()
    pwr = w.getpower()
    
    scalespec = np.sum(pwr, axis=1) / scales
    y = w.fourierwl * scales
    x = np.arange(0, data.size, 1)
    
    p = np.repeat(pwr, 20, axis=0)
    
    plt.figure(figsize=(12,8))
    plt.imshow(p[:1000,:], cmap='cubehelix_r')
    plt.show()
```
